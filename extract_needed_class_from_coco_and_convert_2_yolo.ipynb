```python
##################### consts
COCO_FOLDER    = "/content/data2/"
DARKNET_FOLDER = "/content/darknet2/"
our_coco = COCO(COCO_FOLDER+'annotations/instances_default.json')

cat_names = ['licplate']

##################### vars
class_counter = 0
image_counter = 0

### для перевода в yolov5 format

# image_width  = 0                        # in pixels
# image_height = 0                        # in pixels
# x_center     = 0                        # in pixels
# y_center     = 0                        # in pixels
# bbox_width   = 0                        # in pixels
# bbox_height  = 0                        # in pixels
# x_topleft    = 0                        # in pixels
# y_topleft    = 0                        # in pixels


##################### main
if __name__ == '__main__':
    cat_ids = our_coco.getCatIds(catNms=cat_names)
    train_file = open(DARKNET_FOLDER+"data2/train.txt", "w+")


    for cat_id in cat_ids:
        ann_ids = our_coco.getAnnIds(catIds= cat_id)
        anns = our_coco.loadAnns(ann_ids)

        for ann in anns:
          if len(ann['segmentation']) != 0:
            plate_t_left_x, plate_t_left_y = ann['segmentation'][0][0],ann['segmentation'][0][1]
            plate_t_right_x, plate_t_right_y  = ann['segmentation'][0][2],ann['segmentation'][0][3]
            plate_b_right_x,plate_b_right_y = ann['segmentation'][0][4],ann['segmentation'][0][5]
            plate_b_left_x, plate_b_left_y = ann['segmentation'][0][6],ann['segmentation'][0][7]

            #получение координат для йолов5
            # x_topleft   = ann['bbox'][0]
            # y_topleft   = ann['bbox'][1]
            # bbox_width  = ann['bbox'][2]
            # bbox_height = ann['bbox'][3]

            # print("{}  {}  {}  {}".format(x_topleft, x_topleft, bbox_width, bbox_height))

            img_id = ann['image_id']
            img = our_coco.loadImgs(ids=img_id)
            img_name     = img[0]['file_name']
            
            #перевод координат в формат yolov5
            # image_width  = img[0]['width']
            # image_height = img[0]['height']
            # #x_tl, y_tl, w, h = bbox
            # dw = 1.0 / image_width
            # dh = 1.0 / image_height
            # x_center = x_topleft + bbox_width / 2.0
            # y_center = y_topleft + bbox_height / 2.0
            # x = x_center * dw
            # y = y_center * dh
            # w = bbox_width
            # h = bbox_height
            # w = w * dw
            # h = h * dh
            # print("{}  {}".format(image_width, image_height))
            
            #x_center = x_topleft + bbox_width/2.0
            #y_center = y_topleft + bbox_height/2.0
            #darknet annotation format
            # a = format(x_center/image_width , '.6f')
            # b = format(y_center/image_height , '.6f')
            # c = format(bbox_width/image_width , '.6f')
            # d = format(bbox_height/image_height , '.6f')

            print("{} {} {} {} {} {} {} {} {}".format(class_counter, plate_t_left_x, plate_t_left_y,
                                                       plate_t_right_x, plate_t_right_y,
                                                       plate_b_right_x,plate_b_right_y,
                                                       plate_b_left_x, plate_b_left_y))


            # copy image
            #shutil.copy(COCO_FOLDER + "images/" + img_name, DARKNET_FOLDER + "data/obj/" + img_name)

            # add image name to train.text
            train_file.write("dat2/obj2/{}\n".format(img_name))

            # create text file for annotation
            with open(os.path.splitext(DARKNET_FOLDER + "data2/obj2/" + img_name)[0]+".txt" , "w+") as fp:
              fp.write("{} {} {} {} {} {} {} {} {}\n".format(class_counter, plate_t_left_x, plate_t_left_y,
                                                       plate_t_right_x, plate_t_right_y,
                                                       plate_b_right_x,plate_b_right_y,
                                                       plate_b_left_x, plate_b_left_y))

            image_counter += 1
            print("{}/{} Image .. {}/{} Class\n".format(image_counter, len(anns), class_counter+1, len(cat_names)))

            image_counter = 0
            class_counter += 1
        else:
          pass

        train_file.close()
```
